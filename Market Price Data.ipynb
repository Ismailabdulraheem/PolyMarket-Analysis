```python
import requests
import json
import pandas as pd
```

```python
months = ['february']
clob_token_ids = {}
prices_data = {}

for month in months:
    for i in range(5, 21):
        slug = f'xrp-above-on-{month}-{i}'

        url = f"https://gamma-api.polymarket.com/events/slug/{slug}"
        data = requests.get(url).json()
        markets = data.get('markets', None)
        token_ids = json.loads(markets[0]['clobTokenIds']) if markets else None
        clob_token_ids[slug] = token_ids[0]

        prices_url = f"https://clob.polymarket.com/prices-history?market={clob_token_ids[slug]}&interval=max"
        prices = requests.get(prices_url).json()
        prices_data[slug] = prices.get('history', None)

df = pd.DataFrame.from_dict(clob_token_ids, orient='index', columns=['clob_token_id'])
df['prices'] = df.index.map(prices_data)

print(df)
```
